director analysis

```python
cur.execute("select directors.name,count(movies.original_title),sum(movies.revenue) from directors join movies on directors.id==movies.director_id group by director_id order by sum(revenue) desc")
names=cur.fetchall()
df5=pd.DataFrame(names)
df5
```

```python
cur.execute("select directors.name,count(movies.original_title),sum(movies.revenue) from directors join movies on directors.id==movies.director_id group by director_id order by count(movies.original_title) desc")
names=cur.fetchall()
df5=pd.DataFrame(names)
df5
```
