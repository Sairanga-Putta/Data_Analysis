```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/krishnaik06/Types-Of-Cross-Validation/3c6b95b2f7d6f63b8062d83055f948f948e59a78/cancer_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
x=df.iloc[:,2:]
y=df.iloc[:,1]
```

```python
x.info()
```

```python
x.isna().sum()
```

```python
x.drop(['Unnamed: 32'],axis=1,inplace=True)
```

```python
y
```

```python
x.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=39)
model=LogisticRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
result=model.score(xtest,ytest)
print(result)
```

k_fold


```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
k=KFold(5)
```

```python
model=LogisticRegression()
result=cross_val_score(model,x,y,cv=k)
print(np.mean(result))
```

stratified k fold

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold=StratifiedKFold(n_splits=5)
result=cross_val_score(model,x,y,cv=skfold)
print(np.mean(result))
```

```python
from sklearn.model_selection import LeaveOneOut
lc=LeaveOneOut()
result=cross_val_score(model,x,y,cv=lc)
print(result)
```
